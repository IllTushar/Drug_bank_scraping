```python
salt_code_clean = r'C:\Users\gtush\Desktop\csv_folder\Salt_Code_Cleaner.csv_1.csv'
match_drug_code_file = r'C:\Users\gtush\Desktop\csv_folder\match_drug_file.csv'
# Read CSV files
salt_code_clean_csv = pd.read_csv(salt_code_clean)
match_drug_code_csv = pd.read_csv(match_drug_code_file)
```

```python
desired_similarity=60.0
```

```python
l_distance_list= []
```

```python
for d_name in match_drug_code_csv['drug_name']:
    for names in salt_code_clean_csv['name']:
        gereric_name = salt_code_clean_csv['']
        if d_name == names:
            levien_distance = distance(d_name,names)
            similarity_percentage = (
                                                1 - (levien_distance / max(len(d_name),
                                                                                len(names)))
                                        ) * 100
            l_distance_list.append(similarity_percentage)
```

```python
closer_list = []
```

```python
for distance in l_distance_list:
    if distance >= desired_similarity:
        closer_list.append(distance)
```
